```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the architectural image and the flag image
back_image = cv2.imread('night.jpg')
top_image = cv2.imread('stars.jpg')

points_back= np.array([[0,2791],[4654,2791],[4654,0],[0,0] ], dtype=np.float32)

# Define the four corners of the top image
points_top = np.array([[0, 0], [top_image.shape[1], 0], [top_image.shape[1], top_image.shape[0]], [0, top_image.shape[0]]], dtype=np.float32)

# Compute the homography matrix
homography_matrix, _ = cv2.findHomography(points_top, points_back)

# Warp the flag image to fit the architectural image using the homography matrix
warped = cv2.warpPerspective(top_image, homography_matrix, (back_image.shape[1], back_image.shape[0]))

alpha = 0.5
result = cv2.addWeighted(back_image, 0.7, warped, 0.4, 0)

fig,ax = plt.subplots(1,3,figsize=(15,15))
ax[0].imshow(cv2.cvtColor(back_image, cv2.COLOR_BGR2RGB))
ax[0].set_title('Back Image')
ax[1].imshow(cv2.cvtColor(top_image, cv2.COLOR_BGR2RGB))
ax[1].set_title('Top Image')
ax[2].imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
ax[2].set_title('Blended Image')
plt.show()


```
